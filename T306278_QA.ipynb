June 9 ,2022

```python
query= '''
SELECT event_action, event_context, event_source, count(1) AS events
FROM event.mediawiki_ipinfo_interaction
WHERE year=2022 AND normalized_host.project != 'test'
GROUP BY event_action, event_context, event_source
'''
```

```python
df = hive.run(query)
```

```python
df
```

# Check date

```python
query_date = '''
SELECT normalized_host, min(dt) as start_date, max(dt)
FROM event.mediawiki_ipinfo_interaction
WHERE year=2022 AND normalized_host.project != "test"
GROUP BY normalized_host
ORDER BY start_date, normalized_host
LIMIT 100000
'''
```

```python
df_date = hive.run(query_date)
```

```python
df_date
```

# Check version

```python
query_version ='''
SELECT distinct event_ipinfo_version
FROM event.mediawiki_ipinfo_interaction
WHERE year=2022 AND normalized_host.project != 'test'
LIMIT 100
'''
```

```python
df_v = hive.run(query_version)
```
